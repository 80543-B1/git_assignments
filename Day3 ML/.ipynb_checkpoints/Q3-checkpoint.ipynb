```python
from sklearn.cluster import KMeans
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("universities_std.csv")
df
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df
```

```python
#finding ideal k
wss = []

for i in range(1,11):
    kmeans_model = KMeans(n_clusters=i)
    kmeans_model.fit(df)
    wss.append(kmeans_model.inertia_)

wss
```

```python
x = np.arange(1,11)
plt.plot(x,wss)
```

```python
KMeans_model = KMeans(n_clusters=5)
KMeans_model.fit(df)
```

```python
color = ["Yellow","Red","Orange","Blue","Green"]

for clusters in range(len(KMeans_model.cluster_centers_)):
    plt.scatter(df["SAT"][KMeans_model.labels_ == clusters],
                df["Accept"][KMeans_model.labels_ == clusters],
                color = color[clusters],label = f"cluster {clusters}" )
```
