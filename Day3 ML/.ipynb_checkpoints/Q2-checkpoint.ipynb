```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crime_std.csv")
df
```

```python
df.drop(df.columns[0],axis=1, inplace = True)
```

```python
df
```

```python
wss = []
for k in range(1,11):
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(df)
    wss.append(kmeans_model.inertia_)

wss
```

```python
x = np.arange(1,11)

plt.plot(x,wss)
```

```python
KMeans_model = KMeans(n_clusters = 4)
KMeans_model.fit(df)
```

```python
KMeans_model.labels_
```

```python
colors = ["Red","Green","Blue","Yellow"]
for clusters in range(len(KMeans_model.cluster_centers_)):
    plt.scatter(df["UrbanPop"][KMeans_model.labels_== clusters],
                df["Murder"][KMeans_model.labels_ == clusters],
                color = colors[clusters],label = f"clusters {clusters}")
```
