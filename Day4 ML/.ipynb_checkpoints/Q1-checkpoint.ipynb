## DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
from itertools import product
from sklearn.metrics import silhouette_score
```

```python
eps_values = np.arange(0.5,35.0,0.25)
min_samples = np.arange(1,25)

para = product(eps_values,min_samples)
# print(list(para))
para
```

```python
for (eps_,min_) in para:
    # print(eps_,min_)
    model = DBSCAN(eps = eps_,min_samples = min_)
    model.fit(df)
    
    cluster_count = len(np.unique(model.labels_))

    # silhouette = silhouette_score(df, model.labels_)
    print(f"eps = {eps_}, min_samples = {min_}, count = {cluster_count}")
    
```

```python
for (eps, min_samples) in para:
    # create a dbscan model with all eps and min_sample values
    model = DBSCAN(eps=eps, min_samples=min_samples)

    # fit the data
    model.fit(df)
    
    # find the no of clusters
    cluster_count = len(np.unique(model.labels_))

    # calculate silhouette score
    if cluster_count > 1:
        silhouette = silhouette_score(df, model.labels_)

        # discard the negative values of silhouette score
        # if silhouette < 0:
        print(f"eps = {eps}, min_samples = {min_samples}, count = {cluster_count}, score = {silhouette}")
    else:
        print("cluster count is very small")
```

```python
model = DBSCAN(eps=14.5,min_samples=2)
model.fit(df)
```

```python
colors = ["red", "blue"]

for cluster_index in range(1):
    # plot the data points
    plt.scatter(df['Balance'][model.labels_ == cluster_index], 
                df['Qual_miles'][model.labels_ == cluster_index],
                color=colors[cluster_index], label=f"cluster {cluster_index}")
            
plt.xlabel('Balance')
plt.ylabel('Qual_miles')
# plt.title('Income vs Spending Score')
# plt.legend()
```
